# Part 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

## Functional method

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Multiplots
# Arguments: number of rows, number of columns, plot number you're referring to
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

## Object-oriented method

```python
fig = plt.figure() # Blank canvas

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # (in percentage) Left, bottom, width, height

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.2, 0.4, 0.3])
axes3 = fig.add_axes([0.5, 0.6, 0.4, 0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Larger Plot')

axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

# Part 2

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2)) # width and height

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2)) # width and height

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## Save a figure

```python
# dots per ich, the greater the higher the quality, can be set here or when creating the subplots
fig.savefig('my_picture.png', dpi=200) 
```

## Legends

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc=(0.1, 0.1))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc=10)
```

# Part 3

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='purple', linewidth=3, alpha=0.5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='purple', lw=3, linestyle='--')
```

## Contol over axis appearance

### Plot range

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='purple', lw=2, linestyle='--', marker='o', markersize=10)

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
