```
from pyspark.sql.types import (StructField, StringType,
                               IntegerType, StructType)
```

```
# Use the following when working on Jupyter Notebook
# Changing the schema of a table
# In StructFiled:
# 'age' is the column we are dealing with
# IntegerType is the new type of the 'age' column
# True implies that the column has null values
# data_schema = [StructField('age', IntegerType(), True),
#                StructField('name', StringType(), True)]
# final_struc = StructType(fields=data_schema)
# Then use the following to read your json again and concurrently apply the new data_schema
# df = spark.read.json('people.json', schema=final_struct)
```

```
# Use this when working on Databricks
df = df.withColumn('age', df['age'].cast(IntegerType()))
df = df.withColumn('name', df['name'].cast(StringType()))
```
